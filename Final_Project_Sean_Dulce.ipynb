#Dataset Overview
The dataset "High Income Returns by Income Level and Average Tax Rate" columns consist of the following:


*   **Taxable Year**(Nominal): The year in which the tax data was reported.
*   **Category**(Nominal): Tax category type.
*   **AGIC**(Ordinal): Adjusted Gross Income bracket.
*   **Total Returns**(Ratio): Total number of tax returns filed for a given AGI bracket.
*   **Non-Taxable Returns**(Ratio): Number of returns where no taxes were owed.
*   **Taxable Returns**(Ratio): Number of returns where taxes were owed.
*   **Total Tax Liability**(Ratio): 	Total dollar amount of taxes owed in that income bracket.
*   **Average Tax Rate Less Than 1%**(Ratio): Number of filers in that AGI bracket with less than 1% average tax rate.
*   **Average Tax Rate 1% to 9%**(Ratio): Number of filers in with the tax rate of 1% to 9%.
*   **Average Tax Rate 9% and Above**(Ratio): Number of filers in with the tax rate of 9% and above.

#**Data Preprocessing**

This process helps prepare raw data from the dataset for our analysis.  In order to start this process we must complete 2 methods of Data Preprocessing.

**Cleaning Data** and **Data Aggregation** are the best methods to use for this particular dataset.  

**Cleaning Data** is needed in order to make sure the data has no missing values, or duplicates, in order to reduce the risk of errors when analyzing the data through visualization.  It also ensures that the data is standardized and any white spaces are removed to further ensure aggregation and visualization will be completed accurately.

**Data Aggregation** will be used to group up multiple pieces of data and summarize them into more readable rows of data which can help with the answering any analytical questions and the visualizations.

#Cleaning Data

```python
#Check for missing values in each column
missing_per_column = df.isnull().sum()
print(missing_per_column)

#Check how many total rows have at least one missing value
missing_rows = df.isnull().any(axis=1).sum()
print(f"Total rows with missing values: {missing_rows}")
```

```python
#Check how many duplicate rows exist
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

#View the duplicates
df[df.duplicated()]

#Remove duplicates if any are found
df = df.drop_duplicates()
```

```python
#Strip leading or trailing whitespace in the columns in order to prevent errors when referencing.
df.columns = df.columns.str.strip()
print(df.columns)
```

```python
#Standardizing income labels in the AGIC columns in order to separate the brackets accurately.
df['AGIC'] = df['AGIC'].replace({
    '50,000 to 100,000': '50,001 to 100,000',
    '100,000 to 200,000': '100,001 to 200,000'
})
```

```python
#Convert numeric strings to integers in order to properly calculate the data for any future uses like aggregation or visualization.
columns_to_clean = ['Total Returns', 'Non-Taxable Returns', 'Taxable Returns', 'Total Tax Liability']

for col in columns_to_clean:
    df[col] = df[col].str.replace(',', '').astype(int)
```

#Data Aggregation

```python
#Scaled down 'Total Tax Liability' from cents to dollars to properly perform visualization.
df['Total Tax Liability'] = df['Total Tax Liability'] / 100
```

```python
#Combines rows with the same AGIC value and aggregates numeric data across them.
grouped_df = df.groupby('AGIC')[['Total Returns', 'Non-Taxable Returns', 'Taxable Returns', 'Total Tax Liability']].sum().reset_index()
grouped_df
```

```python
#Groups and divides 'Taxable Returns' with 'Total Returns' in order to get the portion of all filed returns in each bracket resulted in owed taxes.
grouped_df['% Taxable Returns'] = (grouped_df['Taxable Returns'] / grouped_df['Total Returns']) * 100
grouped_df
```

```python
#Groups and divides 'Total Tax Liability' with 'Taxable Returns' in order to measure the average tax paid per person in each bracket.
grouped_df['Avg Tax per Return'] = grouped_df['Total Tax Liability'] / grouped_df['Taxable Returns']
grouped_df
```

```python
#Sorts 'Total Tax Liability' by descending order.
grouped_df = grouped_df.sort_values(by='Total Tax Liability', ascending=False)
grouped_df
```

#**Analytical Questions and Data Visualization**

Now that data has been through preprocessin, now we can look at some analytical questions to analyze this data.

In the previous step, aggregation has been completed to specifically cater to these 3 analytical questions.

#Question 1: What percentage of tax returns are taxable in each income bracket?

Visualization choice: The best visualization for this question is a bar chart as it will be good at comparing percentages across different categories.

```python
#Create the bar chart
plt.figure(figsize=(8, 5))
bars = plt.bar(grouped_df['AGIC'], grouped_df['% Taxable Returns'])

#Add percentage labels above each bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height:.1f}%', ha='center', fontsize=10)

plt.title('Percentage of Taxable Returns by Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('% Taxable Returns')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Insights: Based on our visualization, most taxable returns come from individuals that fall under any of the income brackets that are over 50,000.  On the contrary, of all the individuals in the under 50,000 income bracket, only 43.7% of them have a taxable return.

Of the individuals in the 50,001 to 100,000 income bracket, 92% of them have taxable returns while individals for both of the income brackets that are 100,000 and over are at 99% of taxable returns within the brackets.

#Question 2: How does the average tax paid per taxable return vary across income levels?

Visualization choice: The best visualization for this question is a line graph as it will be good for analyzing the trends at different income levels

```python
#Create line graph
plt.figure(figsize=(10, 6))
plt.plot(grouped_df['AGIC'], grouped_df['Avg Tax per Return'], marker='o', linestyle='-', color='steelblue')

#Annotate with properly scaled numbers
for i, value in enumerate(grouped_df['Avg Tax per Return']):
    plt.text(i, value + 2000, f'${value:,.0f}', ha='center', fontsize=10)

plt.title('Average Tax Paid per Return by Income Bracket')
plt.xlabel('Income Bracket')
plt.ylabel('Avg Tax per Return ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Insights: Based on this visualization, it seems that the income bracket 200,000 and over has a huge disparity compared to the other income brackets with an average of \$219,173 tax paid which is a whopping 90% higher average tax paid than the next income bracket of 100,000 to 200,000.  At the opposite side of the spectrum, the income bracket of 50,000 and under sits at a \$998 average tax paid.

#Question 3: Which income bracket contributes the most to total tax revenue?

Visualization choice: The best visualization for this question is a pie chart as it will be good for showing the proportions of the tax liability levels.

```python
#Create pie chart
plt.figure(figsize=(6, 6))
plt.pie(grouped_df['Total Tax Liability'], labels=grouped_df['AGIC'], autopct='%1.1f%%', startangle=140)
plt.title('Share of Total Tax Liability by Income Bracket')
plt.tight_layout()
plt.show()
```

Insights: Based on this pie chart, it seems that the 200,000 and over income bracket contributes to the most total tax revenue at 75.5%.  The next runner up being 100,000 to 200,000 at a low 15.4% and the next income levels combining for ~9% of the total tax liability.

#Bonus Chart showcasing Total Tax Returns by Income Bracket

```python
#Group returns by bracket
grouped_returns = df.groupby('AGIC')[['Total Returns']].sum().reset_index()

#Sort income brackets in a logical order
income_order = ['Under 50,000', '50,001 to 100,000', '100,001 to 200,000', '200,001 and over']
grouped_returns['AGIC'] = pd.Categorical(grouped_returns['AGIC'], categories=income_order, ordered=True)
grouped_returns = grouped_returns.sort_values('AGIC')

#Create pie chart
plt.figure(figsize=(8, 8))
plt.pie(
    grouped_returns['Total Returns'],
    labels=grouped_returns['AGIC'],
    autopct='%1.1f%%',
    startangle=140
)
plt.title('Share of Total Tax Returns by Income Bracket')
plt.tight_layout()
plt.show()
```
